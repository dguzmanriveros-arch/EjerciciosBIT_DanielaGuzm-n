## Actividad 2: Control de flujo - Clasificación de Estudiantes
Usar bucles y condiciones para clasificar edades y buscar nombres en una lista de estudiantes.

* Parte 1: Bucle for con edades
1. Crea una lista de edades: [12, 17, 8, 15, 22, 9, 30, 25].
2. Usa un bucle for para iterar sobre cada edad:
3. Si la edad es 25, imprime "¡Encontramos al estudiante de 25 años! Deteniendo el análisis." y termina el bucle con break.
4. Si la edad es menor a 10, omite esa iteración con continue.
5. Si la edad es 18 o mayor, imprime "Adulto: [edad] años".
6. En cualquier otro caso, imprime "Menor: [edad] años".

* Parte 2: Bucle while con nombres
1. Crea una lista de nombres: ["Juan", "Pedro", "Jorge", "María","Ana"].
2. Usa un bucle while para iterar sobre los nombres:
3. Si el nombre es "Ana", imprime "¡Ana está en la lista!" y termina el bucle con break.
4. Si el nombre empieza con "J", omite ese nombre con continue.
5. En cualquier otro caso, imprime el nombre.

```python
# PARTE 1
# Punto 1: Creación de edades

edades = [12,17,8,15,22,9,30,25]
print("Serie de edades:")
edades
```

```python
# PARTE 1
# Punto 2: Bucle sobre edades

for estudiante, edad in enumerate(edades):
    print(f"El estudiante {estudiante} tiene una edad de: {edad}")
```

```python
# PARTE 1
# Punto 3: Encontrar estudiante 25 

for estudiante, edad in enumerate(edades):
    print(f"El estudiante {estudiante} tiene una edad de: {edad}")
    if edad == 25:
        print("¡Encontramos al estudiante de 25 años! Deteniendo el análisis.")
        break
```

```python
# PARTE 1
# Punto 4: Omitir edades menores a 10

for estudiante, edad in enumerate(edades):
    if edad < 10:
        continue
    print(f"El estudiante {estudiante} tiene una edad de {edad}")
```

```python
# PARTE 1
# Punto 5 y 6: Adultos 18 o más o menores

for estudiante, edad in enumerate(edades):
    if edad >= 18:
        print(f"El estudiante {estudiante} es un Adulto y tiene {edad} años")
    else:
        print(f"El estudiante {estudiante} es un Menor y tiene {edad} años")
```

```python
# PARTE 2
# Punto 1: Lista de nombre

nombres = ["Juan", "Pedro", "Jorge", "María", "Ana"]
print('Nombres de estudiantes:')
nombres
```

```python
# PARTE 2
# Punto 2: Bucle while para nombres

indice = 0 
listado = len(nombres)
while indice < listado:
    nombre = nombres[indice]
    if nombre is not None:
        print(f'El estudiante {indice} se llama "{nombre}"')
    indice += 1
```

```python
# PARTE 2
# Punto 3: Ana está en la lista

indice = 0 
listado = len(nombres)
while indice < listado:    
    nombre = nombres[indice]
    if nombre == "Ana":
        print(f'¡{nombre} está en la lista, en la posición {indice}!')
        break
    if nombre is not None:
        print(f'El estudiante {indice} se llama "{nombre}"')
    indice += 1
```

```python
# PARTE 2
# Punto 4 y 5: Ana está en la lista

indice = 0 
listado = len(nombres)
while indice < listado:    
    nombre = nombres[indice]
    if nombre is None:
        indice += 1
        continue
    if nombre.startswith("J"):
        indice += 1
        continue 
    print(f'El estudiante {indice} se llama "{nombre}"')
    indice += 1
```
