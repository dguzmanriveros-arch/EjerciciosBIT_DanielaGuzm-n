### 6. Para cada ProductID, suma la cantidad vendida (SUM(OrderQty)). Devuelve solo productos con ventas totales > 1000, ordenados de manera descendente.
Usa GROUP BY, HAVING y ORDER BY

```python
df_sales_detail = pd.read_sql_query("SELECT TOP 2 * FROM SalesLT.SalesOrderDetail", conn)
df_sales_detail
```

```python
#Respuesta punto 6

sql_query = """
SELECT
    ProductID,
    SUM(OrderQty) AS Cant_Total_Vendido,
    SUM(LineTotal) AS Total_Vendido
FROM
    SalesLT.SalesOrderDetail
GROUP BY
    ProductID
HAVING
    SUM(LineTotal) > 1000
ORDER BY
    Cant_Total_Vendido DESC
"""

df_ventas_altas = pd.read_sql_query(sql_query, conn)
df_ventas_altas
```

### 7. Productos sin ventas: lista productos que no aparecen en SalesLT.SalesOrderDetail.
Usa LEFT JOIN + WHERE

```python
#Respuesta punto 7

sql_query = """
SELECT
    P.ProductID,
    P.Name AS ProductName
FROM
    SalesLT.Product AS P
LEFT JOIN
    SalesLT.SalesOrderDetail AS S
ON
    P.ProductID = S.ProductID
WHERE
    S.ProductID IS NULL
"""

df_productos_sin_ventas = pd.read_sql_query(sql_query, conn)
df_productos_sin_ventas
```

### 8. Clientes por país: total de clientes por CountryRegion (usa Address + CustomerAddress).
Podrias necesitar: COUNT(DISTINCT CustomerID).

```python
df_sales_detail = pd.read_sql_query("SELECT TOP 2 * FROM SalesLT.Address", conn)
df_sales_detail
```

```python
df_sales_detail = pd.read_sql_query("SELECT TOP 2 * FROM SalesLT.CustomerAddress", conn)
df_sales_detail
```

```python
#Respuesta punto 8

sql_query = """
SELECT
    A.CountryRegion,
    COUNT(DISTINCT C.CustomerID) AS TotalClientes
FROM
    SalesLT.Customer AS C
INNER JOIN
    SalesLT.CustomerAddress AS CA
    ON C.CustomerID = CA.CustomerID
INNER JOIN
    SalesLT.Address AS A
    ON CA.AddressID = A.AddressID
GROUP BY
    A.CountryRegion
ORDER BY
    TotalClientes DESC
"""

df_clientes_por_pais = pd.read_sql_query(sql_query, conn)
df_clientes_por_pais
```

### 9. Ticket promedio por año: AVG(TotalDue) por YEAR(OrderDate).

```python
df_sales_detail = pd.read_sql_query("SELECT DISTINCT YEAR(OrderDate) FROM SalesLT.SalesOrderHeader", conn)
df_sales_detail
```

```python
#Respuesta punto 9

sql_query = """
SELECT
    YEAR(OrderDate) AS Anio,
    AVG(TotalDue) AS TicketPromedioAnual
FROM
    SalesLT.SalesOrderHeader
GROUP BY
    YEAR(OrderDate)
ORDER BY
    Anio DESC
"""

df_ticket_promedio = pd.read_sql_query(sql_query, conn)
df_ticket_promedio
```
