* Objetivo: Administrar un inventario de una tienda

Inventario inicial:
inventario = [
    {"producto": "camisa", "precio": 25900, "stock": "11"},
    {"producto": "pantalón", "precio": 39900, "stock": "23"},

- Añade un nuevo producto al inventario con las siguientes caracteristicas:
{"producto": "abrigo", "precio": 50000, "stock": "2"}
- Convierte los valores del stock en numeros enteros
- Actualiza el precio de cada producto sumando 10000 al precio original
- Imprime cada producto del inventario y sus caracteristicas con el siguiente formato:
- print(f"Hay {stock} unidades del producto {producto}. Su precio por unidad es de {precio}")


```python

inventario = [
    {"producto": "camisa", "precio": 25900, "stock": "11"},
    {"producto": "pantalón", "precio": 39900, "stock": "23"},
]
print(inventario)
```

```python
#Punto 1: Añadir nuevo producto

inventario.append({
    "producto": "abrigo", "precio": 50000, "stock": "2"}) 
```

```python
inventario
```

```python
#Punto 2: Convertir valores stock en enteros

producto_1 = inventario[0]
producto_1['stock'] = int(producto_1.get("stock"))

producto_2 = inventario[1]
producto_2['stock'] = int(producto_2.get("stock"))

producto_3 = inventario[2]
producto_3['stock'] = int(producto_3.get("stock"))
```

```python
inventario
```

```python
#Punto 3: Sumar 10.000 al precio original
for precio_prod in inventario: 
    precio_prod['precio'] = precio_prod['precio'] + 10000
inventario
```

```python
# Punto 4: Impresión producto por formato 

for impresion in inventario:
    stock = impresion['stock']
    producto = impresion['producto']
    precio = impresion['precio']
    print(f"Hay {stock} unidades del producto {producto}. Su precio por unidad es de {precio}")
```
